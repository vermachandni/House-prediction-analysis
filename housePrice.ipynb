```python
# Compare the difference between all the years features with SalesPrice
# List of year features
year_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']

# Plotting the relationship between year features and SalesPrice
for feature in year_features:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=feature, y='SalePrice')
    plt.title(f'{feature} vs SalePrice')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()

```

```python

if 'SalePrice' in df.columns and 'YearBuilt' in df.columns:
    # Define features to compare
    features = ['MSSubClass', 'LotConfig', 'BldgType', 'YearBuilt']  

    # Loop through each feature and create plots
    for feature in features:
        if feature in df.columns:
            plt.figure(figsize=(14, 8))
            sns.boxplot(x='YearBuilt', y='SalePrice', hue=feature, data=df)
            plt.title(f'Sale Price Distribution by Year Built and {feature}')
            plt.xlabel('Year Built')
            plt.ylabel('Sale Price')
            plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
            plt.legend(title=feature)
            plt.grid(True)
            plt.show()
        else:
            print(f"Feature '{feature}' is not in the dataset.")
else:
    print("Required columns 'SalePrice' or 'YearBuilt' are missing from the dataset.")
```

```python
#On the Discrete Variable Find the relationship between Discrete and Sales price
discrete_vars = ['MSSubClass', 'MSZoning', 'LotConfig', 'BldgType', 'OverallCond']  # Add or modify as needed

# Loop through each discrete variable and create plots
for var in discrete_vars:
    if var in df.columns:
    
        plt.figure(figsize=(14, 8))
        
        # Bar plot showing the average sale price for each category
        plt.subplot(1, 2, 1)
        sns.barplot(x=var, y='SalePrice', data=df, estimator='mean', palette='viridis')
        plt.title(f'Average Sale Price by {var}')
        plt.xlabel(var)
        plt.ylabel('Average Sale Price')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
   
```

```python
#) On the Continous Variable Find the relationship between Discrete and Sales Price

# Define the continuous and discrete variables
continuous_var = 'LotArea'  
discrete_vars = ['MSSubClass', 'LotConfig', 'BldgType'] 

# Check if continuous and discrete variables exist in the dataset
if continuous_var in df.columns:
    for discrete_var in discrete_vars:
        if discrete_var in df.columns:
            plt.figure(figsize=(14, 8))
            
            # Box plot to show the distribution of SalePrice for each category of the discrete variable
            plt.subplot(1, 2, 1)
            sns.boxplot(x=discrete_var, y='SalePrice', data=df, palette='viridis')
            plt.title(f'Sale Price Distribution by {discrete_var}')
            plt.xlabel(discrete_var)
            plt.ylabel('Sale Price')
            plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
            
            # Scatter plot to show the relationship between the continuous variable and SalePrice
            plt.subplot(1, 2, 2)
            sns.scatterplot(x=continuous_var, y='SalePrice', hue=discrete_var, data=df, palette='viridis')
            plt.title(f'Relationship between {continuous_var} and Sale Price by {discrete_var}')
            plt.xlabel(continuous_var)
            plt.ylabel('Sale Price')
            
            plt.tight_layout()
            plt.show()
        
```

```python
#Analyse the Continous values by creating the histogram to understand the distribution.

continuous_vars = ['LotArea', 'TotalBsmtSF', 'SalePrice', 'BsmtFinSF2']  

# Plot histograms for each continuous variable
plt.figure(figsize=(14, 10))  

for i, var in enumerate(continuous_vars, 1):
    plt.subplot(2, 2, i) 
    df[var].dropna().hist(bins=30, edgecolor='black')  # Drop NaN values before plotting
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')

plt.tight_layout() 
plt.show()

```

```python
#Find Continous Features vs Sales Price
continuous_vars = ['LotArea', 'TotalBsmtSF', 'SalePrice', 'BsmtFinSF2']  
# Calculate correlation with SalePrice
correlations = df[continuous_vars].corr()['SalePrice']
print("Correlation with SalePrice:")
print(correlations)

# Plot scatter plots for each continuous variable vs SalePrice
plt.figure(figsize=(14, 10))

for i, var in enumerate(continuous_vars):
    if var != 'SalePrice':  
        plt.subplot(2, 2, i + 1)  # Adjust subplot layout based on the number of variables
        sns.scatterplot(x=df[var], y=df['SalePrice'])
        plt.title(f'{var} vs SalePrice')
        plt.xlabel(var)
        plt.ylabel('SalePrice')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
#Find the Correlation between Numerical Features and Sales Price



# Select only numerical columns
numeric_df = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Extract the correlation with SalePrice
if 'SalePrice' in correlation_matrix.columns:
    correlation_with_salesprice = correlation_matrix['SalePrice'].sort_values(ascending=False)
    print("Correlation with SalePrice:")
    print(correlation_with_salesprice)
else:
    print("SalePrice is not in the numeric columns.")

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

```

```python
#) Find the relationship between Categorical feature and Sales Price

# List of categorical features
categorical_features = ['MSZoning', 'LotConfig', 'BldgType', 'Exterior1st']  
# Analyze each categorical feature
for feature in categorical_features:
    # Calculate mean SalePrice for each category
    mean_saleprice = df.groupby(feature)['SalePrice'].mean()
    print(f"\nMean SalePrice by {feature}:")
    print(mean_saleprice)
    
    # Plot box plot for each categorical feature vs SalePrice
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=feature, y='SalePrice', data=df)
    plt.title(f'SalePrice vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.xticks(rotation=45)  # Rotate x-axis labels if needed
    plt.show()

    
```
